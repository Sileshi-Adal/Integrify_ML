## What can be done in Pandas 

Pandas is a popular Python package for data science, and with good reason: it offers powerful, expressive and flexible data structures that make data manipulation and analysis easy.

- DataFrame is very useful structure to manipulate datas in pandas. Data frame is a way to store data in rectangular grids that can easily be overviewed.
- You can specify the index and column names for your DataFrame. It is handy way to manipulate your data
- with describe() command you can see a quick statistic summary of your data
- Boolean indexing allows you to find necessary points 
- with one command -dataFrame.dropna()- data can be cleaned from rows with any column having null data 
- matplotlib package that is used to visualize data is usable in pandas
- .apply() is used to apply a function to all columns
- .map property applies changes to every element of a column. It is similar with .apply but you can do all changes in one line with .map
- .unique() returns unique entries in selected columns. It allows us to see unique elements in a column
- merging dataframes: essential way to collate informations come from different sources
- pandas has easy sorting applications
- Below commands are used to see summary of spesific data:
    - .sum : sum of values in a dataframe
    - .max : maximum value of a dataframe
    - .min : minimum value of a dataframe
    - .mean : average value of a dataframe
    - .median : median value of a dataframe

## Cleaning

1. Drop all the missing rows which contain any column with missing data

```python
## 1 dropping all the missing rows which contain any missing data 
df = pd.DataFrame(movie_data)
df.dropna(inplace=True)
df.head()
```

4. Convert the duration column to numerical representation of seconds

```python
df.Duration = df["Duration"].map(lambda x: int(x.replace("mins."," "))*60)
df.head()
```

3. Remove the year from title column

```python
df.Title=df.Title.apply(lambda x: x[:-6])
df.head()
```

##  Concatenating 

1. Read crew_data.tsv.gz to DataFrame named movie_crew_df with columns ID, director_ids indexed by ID (forget the rest of the columns)

```python
movie_crew_df = pd.read_csv('crew_data.tsv.gz', delimiter=('\t'),names=['ID', 'director_ids'],index_col="ID",usecols=[0,1],header=0)

movie_crew_df.head()
```

2. Read person_data.tsv.gz to DataFrame named person_df with columns person_ID, name indexed by person_ID (forget the rest of the columns)

```python
person_df = pd.read_csv('name_data.tsv.gz', delimiter=('\t'),names=['person_ID', 'name'],usecols=[0,1],index_col=0,header=0)
```

```python
person_df.head()
```

```python
person_df=person_df.reset_index()
```

```python
person_df.head()
```

3. Merge movie_crew_df into df with IDs and drop the columns without necessary data (hint outer-join)

```python
## pd.merge()
df=pd.merge(df,movie_crew_df,left_on="ID",right_on="ID",how="inner")    
df.head()
```

4.Remove movies with more than one director and rename director_ids => director_id


```python
df=df.rename(columns={'director_ids':'director_id'})
df=df.loc[df["director_id"].str.len()==9]
df.head()
```

5. Merge df with person_df on director_id, remove director_id and person_ID columns, rename name => director

```python
merge_df_person_df=pd.merge(df,person_df,left_on="director_id",right_on="person_ID",how="inner")
```

```python
merge_df_person_df=merge_df_person_df.drop(['director_id', 'person_ID'], axis=1)
```

```python
merge_df_person_df.head()
```

```python
merge_df_person_df=merge_df_person_df.rename(columns={'name':'director'})
```

```python
merge_df_person_df.head()
```
