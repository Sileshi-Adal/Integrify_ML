### Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
url= 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
data = pd.read_csv(url, index_col=0, delimiter = ('|'))
```

### Step 3. Assign it to a variable called users.

```python
users=data
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
mean_age_per_occupation=users.groupby('occupation').agg({'age':'mean'})
mean_age_per_occupation.head()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
male_ratio_per_occupation=users.groupby(['occupation', 'gender']).size().unstack().assign(male_ratio=lambda x: x.M / x.sum(axis=1))
male_ratio_per_occupation.head()
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
minimum_ages=users.groupby('occupation').agg({'age':'min'}) 
maximum_ages=users.groupby('occupation').agg({'age':'max'})
minimum_ages=minimum_ages.rename(columns={'age':'min_age'})
maximum_ages=maximum_ages.rename(columns={'age':'max_age'})
```

```python
maximum_minimum_ages=pd.concat([minimum_ages, maximum_ages], axis=1)
maximum_minimum_ages.head()
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
mean_age_for_com_occu=users.groupby(["occupation","gender"])["age"].mean()
mean_age_for_com_occu.head()
```

### Step 8.  For each occupation present the percentage of women and men

```python
percentage_data = pd.DataFrame({'Percentage': users.groupby(('occupation', 'gender')).size() / len(users)})
```
